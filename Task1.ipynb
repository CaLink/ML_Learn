##Типизация
Присвойте соответствующим колонкам один из типов (category, float, int)

```python
df = df.drop(['Rape (legacy definition)'], axis=1)
for col in df.columns:
  if col == 'City':
    continue
  df[col] = df[col].astype(str).str.replace(',','')
for col in df.columns:
  if col == 'City':
    df[col].astype("category")
    continue
  df[col] = df[col].astype('Float64')#Вывести типы колонок тут
```

##Задание для гениев *
Агрегируйте числовой признак по категориальному признаку с использованием метода **без использования** библиотеки pandas, и вычислите медиану и среднее значение для каждой категории.

```python
from numpy.core.fromnumeric import mean
#Агрегация
temp = df[['City','Population']].values.tolist()

agro = {}

for row in temp:
  if row[0] not in agro.keys():
    agro[row[0]] = []
  agro[row[0]].append(row[1])

for row in agro.keys():
  print('Среднее {:18s} {:10.2f}'.format(row, np.mean(agro[row])))
  print('Медиана {:18s} {:10.2f}'.format(row, np.median(agro[row])))
```

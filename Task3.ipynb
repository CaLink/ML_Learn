````python
import pandas as pd

df = pd.read_csv('ca_offenses_by_city.csv')
````

````python
# чиним датасет

df = df.drop(['Rape (legacy definition)'], axis=1)
for col in df.columns:
  if col == 'City':
    continue
  df[col] = df[col].astype(str).str.replace(',','')
for col in df.columns:
  if col == 'City':
    df[col].astype("category")
    continue
  df[col] = df[col].astype('int')
````

````python
#Ломаем данные
broken_df = df.mask(np.random.random(df.shape)<0.1)
broken_df
````

````python
#Удаление
droped_df = broken_df.dropna()

# Мы просто удалили все строчки, где содержались nan'ы, не разбирая их природу
# Данный способ хорош, когда мы точно знаем, что мы не можем обработать запись без какого-то поля
# Но количество nan'ов в выборке может быть слишком большим. Настолько, что после удаления выборка станет не репрезентативной

droped_df
````

````python
#Заполнение при помощи медианы
filled_median_df = broken_df.fillna(broken_df.median().astype(int))
filled_median_df['City'] = broken_df.fillna(broken_df.mode())['City']

# Данный способ позволяет в добавить значения в пропущенные ячейки
# Добавляя в них медиану, не изменным останеться параметр медианы
# Считаю его более приемлимым, т.к. медиана меньше подвержена выбросам
# После применения медианы/среднего работа с модой станет некорректной
#
# Однако заполнение медианой/средним не сработает с категориальными переменными
# Поэтому применим к ним [City] моду

filled_median_df
````

````python
#Заполнение при помощи среднего

#Заполнение при помощи медианы
filled_mean_df = broken_df.fillna(broken_df.mean().astype(int))
filled_mean_df['City'] = broken_df.fillna(broken_df.mode())['City']

# Делает все тоже самое, только заполняем средним
# Среднее столбца останеться неизменным


filled_mean_df
````

````python
#Заполнение при помощи моды

filled_mode_df = broken_df.fillna(broken_df['Robbery'].mode()[0])

# Заполняем модой все нули. Изменилась логика использования функции
# Хороша в категориальных переменных
# Но плоха в количественных

filled_mode_df
````

````python
#Заполнение при помощи интерполяции

interpolated_df = broken_df.interpolate()

# Заполняем данные на основе ближайших соседей
# Ломается, если использовать 64-битные типы данных
# По сути является предсказыванием. Поэтому медиана, среднее и мода изменяться
# Опять же, не работает с категориальными
# Плохо сработает с данными в начале и конце, т.к. взять их можно только с одной стороны

interpolated_df
````

##Поиск выбросов и аномалий

**Те, кто выполняет задания со * делают поиск выбросов с помощью написанного ранее класса**

Найти выбросы по числовой переменной (использовать метод основанный на среднем и на межквартильном размахе). Объясните результат

Вывести статистику с выбросами и без. Сравните и объясните результат


````python
# @title ИМПОРТ СУПЕР КРУТОЙ ЛИБЫ

from sklearn.cluster import DBSCAN
import numpy as np
import math

class Outline:
  def __init__(self, df):
    self.df = df

  def by_mean_and_std(self, col):

    outlier_index_list = []

    index_list = list(df.index)
    arr = list(df[col])
    n = len(arr)
    mean = np.array(arr).mean()
    median = np.array(arr).mean()
    std = np.array(arr).std()

    close_left = abs(median-mean)
    close_right = 3*((std*std/n)**0.5)

    cof = 3 if close_left <=close_right else 5

    for i in range(df[col].shape[0]):

      arr = list(df[col])
      x = arr.pop(i)
      mean = np.array(arr).mean()
      std = np.array(arr).std()

      lmin = mean-(cof*std)
      lmax = mean+(cof*std)

      if (x > lmax) | (x < lmin):
        outlier_index_list.append(index_list[i])

    ans = df[~df.index.isin(outlier_index_list)]
    return ans

  def by_chauvenet(self, col):

    value_list = list(df[col])
    index_list = list(df.index)

    outlier_find = True

    while outlier_find:
      outlier_find = False

      right = 1/(2*len(value_list))
      mean = np.array(value_list).mean()
      std = np.array(value_list).std()

      for i in range(len(value_list)):
        x = abs(value_list[i]-mean)/std
        if math.erfc(x) < right:
          value_list.pop(i)
          index_list.pop(i)
          outlier_find = True
          break

    ans = df.iloc[index_list]
    return ans

  def by_quart(self, col):
    Ni = 1.5
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3-q1

    lmax = q3 + Ni*iqr
    lmin = q1 - Ni*iqr

    ans = df[~((df[col]< lmin) | (df[col]>lmax))]
    return ans

  def by_two_value(self, col1, col2):

    df_selected_colum_values = df[['Violent crime', 'Robbery']].values
    array_of_selected_colum = np.array(df_selected_colum_values)

    clustering = DBSCAN(eps=3, min_samples=2).fit(array_of_selected_colum)

    df_mutable = df
    df_mutable['Cluster'] = list(clustering.labels_)

    grouped_series = df_mutable.groupby('Cluster')['Cluster'].count()
    outlier_cluster = list(grouped_series[grouped_series <= 2].keys())

    ans = df_mutable[~df_mutable['Cluster'].isin(outlier_cluster)]
    return ans



````

Реализовать расчет корреляции Спирмана

````python
#Функция по расчету тут

# Вспомогательная функция для ранжирования
def range_list(x):
  x_sorted = sorted(x)
  dx_dict = {}

  for i in x:
    dx_dict[i] = ''

  dx_dict[x_sorted[0]] = 1
  count_same_x = 0

  for i in range(1, len(x)):
    if x_sorted[i-1] == x_sorted[i]:
      count_same_x += 1
      continue

    if count_same_x > 0:
      sum = (i-count_same_x + i)*(count_same_x+1)/2
      val = sum/(count_same_x+1)
      dx_dict[x_sorted[i-1]] = round(val,2)

      count_same_x = 0

    dx_dict[x_sorted[i]] = i+1

  if count_same_x > 0:
    sum = (i+1-count_same_x + i+1)*(count_same_x+1)/2
    val = sum/(count_same_x+1)
    dx_dict[x_sorted[i-1]] = round(val,2)

  ans = []
  for i in x:
    ans.append(dx_dict[i])

  return(ans)


def spearman_correl(x, y):
  if len(x) != len(y) and len(x) < 2:
    return

  x = list(x)
  y = list(y)

  dx = range_list(x)
  dy = range_list(y)

  n = len(x)
  sum = 0
  for i in range(n):
    sum += (dx[i] - dy[i])*(dx[i] - dy[i])

  ans = 1 - (6*sum)/(n*n*n - n)

  return ans
````

````python
# Передаем методу две колонки
# Сначала расчитаем для каждого значения его порядковый номер
#   Если значения одинаковые, то найдем среднее среди их порядковых номеров
# Затем ищем сумму квадратов разности между рангами и высчитываем оставшуюся часть формулы
# Последним шагом будет интерпритация полученного результата
#   Чем ближе к нулю, тем менее связанны данные и наоборот
#   Чем ближе к -1/1, тем сильнее связанны данные

print(spearman_correl(df['Robbery'], df['Burglary']))
````
